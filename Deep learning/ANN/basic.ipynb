```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as snb 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
x ,y =load_iris(return_X_y=True)
x=pd.DataFrame(x)
y=pd.DataFrame(y)


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x
```

```python
y
```

```python
import keras
from keras.models import Sequential # model that is used to create a neural network
from keras.layers import Dense # layer that is used to create a dense layer
```

```python
object =Sequential() # create a sequential model
object.add(Dense(units=2,kernel_initializer='uniform',activation='relu',input_dim=4)) # add a dense layer to the model, kernel_initializer is used to initialize the weights of the layer , units is the number of units in the layer, activation is the activation function, input_dim is the number of input features
 # that is used in input layer
object.add(Dense(units=2,kernel_initializer='uniform',activation='relu')) # add a dense layer to the model, kernel_initializer is used to initialize the weights of the layer, units is the number of units in the layer, activation is the activation function 
# that is used in hidden layer
object.add(Dense(units=2,kernel_initializer='uniform',activation='relu')) # add a dense layer to the model, kernel_initializer is used to initialize the weights of the layer, units is the number of units in the layer, activation is the activation function ,relu is used for activation function which is used for binary classification and that is used in middle layer 
# that is used in hidden layer
object.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))# add a dense layer to the model, kernel_initializer is used to initialize the weights of the layer, units is the number of units in the layer, activation is the activation function , sigmoid is used for binary classification 
# that is used in output layer
```

```python
object.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) # compile the model, optimizer is used to optimize the model, loss is used to calculate the loss, metrics is used to calculate the metrics , binary_crossentropy is used for binary classification and adam is used for optimization , accuracy is used to calculate the accuracy and loss is used to calculate the loss , binary_crossentropy is used for binary classification , metrics is used to calculate the metrics of a model and accuracy is used to calculate the accuracy of a model
```

```python
history=object.fit(x_train,y_train,batch_size=10,epochs=10) # fit the model, batch_size is used for training, epochs is used for training , batch_size is used for training, epochs is used for training 
```
