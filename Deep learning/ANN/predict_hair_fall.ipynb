```python
cat_col = df.select_dtypes(include = 'O')
num_col = df.select_dtypes(exclude = "O")
```

```python
cat_col
```

```python
from sklearn.preprocessing import OneHotEncoder , OrdinalEncoder
```

```python
encoder = OrdinalEncoder()
```

```python
encoder.fit(cat_col)
```

```python
transform_df = encoder.transform(cat_col)
```

```python
encoder.get_feature_names_out()
```

```python
cat_col = pd.DataFrame(transform_df , columns = encoder.get_feature_names_out())
```

```python
transform_df.shape
```

```python
final_df = pd.concat([ num_col,cat_col ] , axis = 1)
```

```python
final_df
```

```python
final_df.shape
```

```python
final_df.dtypes
```

```python
x = final_df.drop("Hair Loss" , axis = 1)
```

```python
y = final_df[['Hair Loss']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y , test_size = 0.2 , random_state = 42)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.losses import categorical_crossentropy
from keras.metrics import categorical_accuracy
```

```python
x_train.shape[1]
```

```python
object =Sequential() # create a sequential model
object.add(Dense(units=6,kernel_initializer='uniform',activation='relu',input_dim=12)) # add a dense layer to the model, kernel_initializer is used to initialize the weights of the layer , units is the number of units in the layer, activation is the activation function, input_dim is the number of input features
 # that is used in input layer
object.add(Dense(units=6,kernel_initializer='uniform',activation='relu')) # add a dense layer to the model, kernel_initializer is used to initialize the weights of the layer, units is the number of units in the layer, activation is the activation function 
# that is used in hidden layer
object.add(Dense(units=6,kernel_initializer='uniform',activation='relu')) # add a dense layer to the model, kernel_initializer is used to initialize the weights of the layer, units is the number of units in the layer, activation is the activation function ,relu is used for activation function which is used for binary classification and that is used in middle layer 
# that is used in hidden layer
object.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))# add a dense layer to the model, kernel_initializer is used to initialize the weights of the layer, units is the number of units in the layer, activation is the activation function , sigmoid is used for binary classification 
# that is used in output layer
```

```python
object.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) # compile the model, optimizer is used to optimize the model, loss is used to calculate the loss, metrics is used to calculate the metrics , binary_crossentropy is used for binary classification and adam is used for optimization , accuracy is used to calculate the accuracy and loss is used to calculate the loss , binary_crossentropy is used for binary classification , metrics is used to calculate the metrics of a model and accuracy is used to calculate the accuracy of a model
```

```python
history=object.fit(x_train,y_train,batch_size=10,epochs=10) # fit the model, batch_size is used for training, epochs is used for training , batch_size is used for training, epochs is used for training 
```
