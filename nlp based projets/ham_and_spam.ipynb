```python
# trainig test split 
import sklearn 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test=train_test_split(dataset.v2,dataset.spam ,test_size=0.25)
```

```python
x_train.shape
```

```python
x_train[:4]
```

##### create bag to words 

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v =CountVectorizer()
```

```python
x_train_cv=v.fit_transform(x_train.values)
x_test_cv = v.transform(x_test.values)
```

```python
x_train_cv.toarray()[:4][0]
```

```python
x_train_cv.shape
```

```python
v.get_feature_names_out()[1712]
```

```python
v.vocabulary_
```

```python
x_train_np=x_train_cv.toarray()
```

```python
x_train_np[0]
```

```python
import numpy as np 
```

```python
np.where(x_train_np[0]!=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
obj =MultinomialNB()
```

```python
obj.fit(x_train_np,y_train)
```

```python
obj.score(x_train_np,y_train)
```

```python
x_test.shape , x_train.shape, y_test.shape , y_train.shape 
```

```python
pred=obj.predict(x_test_cv)
```

```python
print(pred)
```

```python
from sklearn.metrics import classification_report
```

```python
output =classification_report(y_test,pred)
```

```python
print(output)
```
